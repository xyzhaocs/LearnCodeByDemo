```python
# 

import numpy as np
import matplotlib.pyplot as plt

N = 10000
X = np.random.randn(N) * np.sqrt(10) + 5      # X ~ N(5, 10)
Y = np.random.randn(N) * np.sqrt(1) + 9       # Y ~ N(9, 1)

Z = X + Y

min0 = min(np.min(X), np.min(Y), np.min(Z))
max0 = max(np.max(X), np.max(Y), np.max(Z))

plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.hist(X, bins=500)
plt.title(f'X: mean={np.mean(X):.2f}, variance={np.var(X):.2f}')
plt.xlim([min0, max0])
plt.ylim([0, 100])

plt.subplot(3, 1, 2)
plt.hist(Y, bins=500)
plt.title(f'Y: mean={np.mean(Y):.2f}, variance={np.var(Y):.2f}')
plt.xlim([min0, max0])
plt.ylim([0, 100])

plt.subplot(3, 1, 3)
plt.hist(Z, bins=500)
plt.title(f'Z: mean={np.mean(Z):.2f}, variance={np.var(Z):.2f}')
plt.xlim([min0, max0])
plt.ylim([0, 100])

plt.tight_layout()
plt.show()
```
