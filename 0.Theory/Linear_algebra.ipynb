# 协方差 Covariance

```python
import numpy as np

# 生成两个随机变量 X 和 Y
np.random.seed(0)  # 设置随机种子以确保结果可重复
X = np.random.normal(loc=0, scale=1, size=100)  # 均值为0，标准差为1的正态分布
Y = 2 * X + np.random.normal(loc=0, scale=0.5, size=100)  # Y 是 X 的线性变换加上噪声

# 计算协方差矩阵
cov_matrix = np.cov(X, Y)

# 输出协方差矩阵
print("协方差矩阵:")
print(cov_matrix)

# 提取 X 和 Y 之间的协方差
cov_XY = cov_matrix[0, 1]
print(f"X 和 Y 之间的协方差: {cov_XY}")
```

```python
print(sum((X[i]-np.mean(X))*(Y[i]-np.mean(Y)) for i in range(len(X)))/(len(X)-1) )
print(np.var(X)*100/99)
print(cov_matrix[0,0])
```
